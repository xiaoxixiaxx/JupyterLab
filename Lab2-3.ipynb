```python
import geopandas as gpd
myMap = gpd.read_file(r'Data/world_xia.geojson')
%matplotlib inline
myMap.plot(color='purple', figsize=(16,16))
```

Explaination

The first line I wrote imports the GeoPandas library and stores its name as 'gpd'. 
The second line is using 'read_file' command to read the data and assigning it to a dataframe variable 'myMap', which refers to a pandas dataframe.
The third line uses the plotting capabilities of the matplotlib library.
The last line polts a map of 'myMap' which bourndary by 'world_xia.geojson' file. I set the color as purple and size is 16 by 16 inches.
